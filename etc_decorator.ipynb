### Decorator(데코레이터)
- 클래스에서 메서드를 만들 때, *데코레이터1* 식으로 @로 시작하는 것들
- 함수(메서드)를 장식한다고 해서 데코레이터라는 이름이 붙여짐

```python
class Calc :
    @staticmethod 
    def add(a, b) :
        print(a + b)
```

### 데코레이터 만들기
- 함수를 수정하지 않은 상태에서 추가 기능을 구현할 때 사용

```python
def hello() :
    print('hello 함수 시작')
    print('hello')
    print('hello 함수 끝')

def world() :
    print('world 함수 시작')
    print('world')
    print("world 함수 끝")

hello()
world()
```

```python
# 함수의 시작과 끝은 출력하는 데코레이터
def trace(func) :
    def wrapper() :
        print(func.__name__, '함수 시작')  # 함수 이름 출력
        func() 
        print(func.__name__, '함수 끝')
    return wrapper

def hello() :
    print('hello')
def world() :
    print('world')

# 데코레이터에 호출할 함수 넣음
trace_hello = trace(hello) 
trace_hello()
trace_world = trace(world)
trace_world()
```

```python
# @로 데코레이터 사용하기
# 함수의 시작과 끝은 출력하는 데코레이터
def trace(func) :
    def wrapper() :
        print(func.__name__, '함수 시작')  # 함수 이름 출력
        func() 
        print(func.__name__, '함수 끝')
    return wrapper

@trace 
def hello() :
    print('hello')

@trace
def world() :
    print('world')

hello()
world()
```
